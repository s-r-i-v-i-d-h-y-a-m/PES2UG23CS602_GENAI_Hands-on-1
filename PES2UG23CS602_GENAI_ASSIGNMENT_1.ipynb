```python
from transformers import pipeline
```

```python
bert = "bert-base-uncased"
roberta = "roberta-base"
bart = "facebook/bart-base"
```

```python
print("EXPERIMENT 1 – TEXT GENERATION")
print("Model: BERT")

generator = pipeline("text-generation", model="bert-base-uncased")
output = generator("The future of Artificial Intelligence is", max_length=30)

print(output)
```

```python
print("EXPERIMENT 1 – TEXT GENERATION")
print("Model: RoBERTa")

generator = pipeline("text-generation", model="roberta-base")
output = generator("The future of Artificial Intelligence is", max_length=30)

print(output)
```

```python
print("EXPERIMENT 1 – TEXT GENERATION")
print("Model: BART")

generator = pipeline("text-generation", model="facebook/bart-base")
output = generator("The future of Artificial Intelligence is", max_length=30)

print(output)
```

```python
print("EXPERIMENT 2 – FILL MASK")
print("Model: BERT")

fill_mask = pipeline("fill-mask", model="bert-base-uncased")
output = fill_mask("The goal of Generative AI is to [MASK] new content.")

print(output)
```

```python
print("EXPERIMENT 2 – FILL MASK")
print("Model: RoBERTa")

fill_mask = pipeline("fill-mask", model="roberta-base")
output = fill_mask("The goal of Generative AI is to <mask> new content.")

print(output)
```

```python
print("EXPERIMENT 2 – FILL MASK")
print("Model: BART")

fill_mask = pipeline("fill-mask", model="facebook/bart-base")
output = fill_mask("The goal of Generative AI is to <mask> new content.")

print(output)
```

```python
print("EXPERIMENT 3 – QUESTION ANSWERING")
print("Model: BERT")

qa = pipeline("question-answering", model="bert-base-uncased")

output = qa(
    question="What are the risks?",
    context="Generative AI poses significant risks such as hallucinations, bias, and deepfakes."
)

print(output)
```

```python
print("EXPERIMENT 3 – QUESTION ANSWERING")
print("Model: RoBERTa")

qa = pipeline("question-answering", model="roberta-base")

output = qa(
    question="What are the risks?",
    context="Generative AI poses significant risks such as hallucinations, bias, and deepfakes."
)

print(output)
```

```python
print("EXPERIMENT 3 – QUESTION ANSWERING")
print("Model: BART")

qa = pipeline("question-answering", model="facebook/bart-base")

output = qa(
    question="What are the risks?",
    context="Generative AI poses significant risks such as hallucinations, bias, and deepfakes."
)

print(output)
```
